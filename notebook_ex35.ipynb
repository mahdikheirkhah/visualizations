The plot has to contain:

- 6 subplots: 2 rows, 3 columns
- Keep space between plots: `hspace=0.5` and `wspace=0.5`
- Each plot contains
  - Text (2,3,i) centered at 0.5, 0.5. _Hint_: check the parameter `ha` of `text`
  - a title: Title i

```python
rows = 2
cols = 3

for i in range(1, 7):
    plt.subplot(rows, cols, i)
    
    plt.xlim(0.0, 1.0)
    plt.ylim(0.0, 1.0)
    
    plt.text(0.5, 0.5, f'(2, 3, {i})', fontsize=18, ha='center', va='center')
    plt.title(f'Title {i}')

plt.subplots_adjust(hspace=0.5, wspace=0.5)

plt.show()
```
