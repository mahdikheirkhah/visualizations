```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

```python
y0 = np.random.randn(50)
y1 = np.random.randn(50) + 1 # shift mean
y2 = np.random.randn(50) + 2
```

1. In the same figure, plot the 3 distributions as box plots as shown in the image. In this exercise the style is not important.

![alt text][logo_ex7]

[logo_ex7]: ./w1day03_ex7_plot1.png "Box plot ex7"

The plot has to contain:

- The title
- The legend

```python
fig = go.Figure()

# Add the three box plot traces
fig.add_trace(go.Box(y=y0, name='Sample A'))
fig.add_trace(go.Box(y=y1, name='Sample B'))
fig.add_trace(go.Box(y=y2, name='Sample C'))

# Add title and ensure legend is visible
fig.update_layout(
    title='Box Plot Comparison',
    xaxis_title='Samples',
    yaxis_title='Values'
)

fig.show()
```
