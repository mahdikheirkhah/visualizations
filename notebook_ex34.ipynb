```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
left_data = [5, 7, 11, 13, 17]
right_data = [0.1, 0.2, 0.4, 0.8, -1.6]
x_axis = [0.0, 1.0, 2.0, 3.0, 4.0]
```

The plot has to contain:

- the title
- name on left y-axis and right y-axis
- **style**:
  - left data in black
  - right data in red

```python
fig, ax1 = plt.subplots()
ax1.plot(x_axis, left_data, color='black')
ax1.set_ylabel('Big')
ax1.set_title('Twin axes')

ax2 = ax1.twinx()

ax2.plot(x_axis, right_data, color='red')
ax2.set_ylabel('Small')

plt.show()
```
